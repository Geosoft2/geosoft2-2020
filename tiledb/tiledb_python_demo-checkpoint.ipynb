Here we open a previously created TileDB array from geospatial.

```python
with tiledb.open("C:\\GIS\\data\\geodemo") as A:
    data = A[1]
    schema = A.schema
data
```

Convert the values to a numpy array so that we can displayed as an image.

```python
npdata = np.asarray(data['TDB_VALUES'], dtype= np.uint8)
```
